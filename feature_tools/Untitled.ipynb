```python
data = ft.demo.load_mock_customer()
```

```python
customer_df = data['customers']
```

```python
customer_df.head(2)
```

```python
sessions_df = data["sessions"]
```

```python
sessions_df.head(2)
```

```python
transactions_df = data["transactions"]

transactions_df.head(2)
```

# 包含数据集中所有实体的字典

```python
entities = {
    'customers': (customer_df, "customer_id"),
    "sessions": (sessions_df, "session_id", "session_start"),
    "transactions": (transactions_df, "transaction_id", "transaction_time")
}
```

```python
relationships = [
    ("sessions", "session_id", "transactions", "session_id"),
    ("customers", "customer_id", "sessions", "customer_id")
]
```

# 运行深度特征合成

dfs：

目的：

有更多新功能来描述用户的行为

方法：

输入：

最小输入：一组实体、一组关系 + ”target_entity“

输出：特征矩阵和特征定义列表

```python
feature_matrix_customers, feature_defs = ft.dfs(entities=entities, 
                                                relationships= relationships,
                                                target_entity="customers")
```

```python
feature_matrix_customers
```

```python
feature_matrix_sessions, feature_defs = ft.dfs(entities=entities, 
                                                relationships= relationships,
                                                target_entity="sessions")
```
